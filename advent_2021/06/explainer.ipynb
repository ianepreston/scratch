```python
import pandas as pd
```

```python
eg_in = [3, 4, 3, 1, 2]
```

```python
cols = [i for i in range(9)]
df = pd.DataFrame(columns=cols)
for col in cols:
    df.loc[0, col] = 0
for x in eg_in:
    df.loc[0, x] += 1
```

```python
days = 18
for i in range(days):
    df.loc[i + 1] = df.loc[i]
    df.loc[i + 1, (i + 7) % 9] += df.loc[i, i % 9]
```

```python
def highlights(x):
    sdf = x.copy()
    sdf.loc[:, :] = 'background-color: white'
    for i in range(days):
        sdf.loc[i, i % 9] = 'background-color: blue'
        sdf.loc[i + 1, (i + 7) % 9] = 'background-color: green'
    return sdf
```

```python
df.style.apply(highlights, axis=None)
```
